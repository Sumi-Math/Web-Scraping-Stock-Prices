```python
strt_date = datetime.datetime(2020,10,15)     #datetime(year,month,day)#datetime function format
print(strt_date)
end_date = datetime.datetime.today()
```

```python
stk_price=pd.DataFrame()
```

```python
comp_energy['Symbol'].unique()
```

```python
yf.download('APA',start=strt_date, end=end_date)



```

```python

for i in comp_energy['Symbol'].unique():
    df = yf.download(i, start=strt_date, end=end_date)
        
    df['Symbol'] = i
    
    stk_price = stk_price.append(df)
    print(df.head())
    
    
    
```
